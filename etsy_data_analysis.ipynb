```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns

etsydata = pd.read_csv('etsy-mens-t-shirts-11-15-2017.csv')
# a good way to display dataframe
display(etsydata.head(10))
```

```python
etsydata.describe()
```

```python
    # Quel tee-shirt est le plus chèr?

# Method 1 : Boolean indexing
# Masque  sur la colonne "product_price" avec conditions (max de la liste etsydata["product_price"])
most_expensive = etsydata[etsydata["product_price"]==max(etsydata["product_price"])]
# afiche la dataframe most_expensive
display(most_expensive)
# affiche le lien (clickable)
print(most_expensive["url"])
```

```python
# Method 2 (Better): loc accessor
# To extract the row with the maximum value in a specific column of a pandas DataFrame, 
# we can use the loc accessor with a boolean mask.
# .loc is a label-based indexing method for pandas DataFrames and Series. 
# It is used to access and modify data based on the label of the rows and columns.
# more concise and efficient as it avoids creating a Boolean mask and filtering the DataFrame.
max_price_row = etsydata.loc[etsydata['product_price'].idxmax()]
display(max_price_row)
```

```python
# Quel tee-shirt a la meilleure review?
best_review = etsydata.loc[etsydata['avg_review'].idxmax()]
display(best_review)
```

```python
# display number of lines & columns
etsydata.shape
```

```python
# Split the dataframe into two parts based on the price

low_price = etsydata[etsydata['product_price'] <= 20 ]
high_price = etsydata[etsydata['product_price'] > 20 ]
display(low_price)
```

```python
# join the two parts back together
concat_df = pd.concat([low_price, high_price])
# checking that the numbers of row and columns are the same as the original
concat_df.shape
```

```python
# select all data with a avg_review equal to NaN
# display all null values .isna()
# select all id listing
# inverse = notna
null_review = etsydata[etsydata["avg_review"].isna() == True]
listingid_w_null_review = null_review['listing_id'].tolist()
print(listingid_w_null_review)
```

```python
    # What kinds of T-Shirts gets the most search traffic (views)?
# sort the dataframe by view count in descending order
sorted_data = etsydata.sort_values('views', ascending=False)

# get the product with the highest view count
most_popular_tshirt = sorted_data.iloc[0]['url']

display(most_popular_tshirt)
```

```python
    # Demonstrate how the price affects the number of views
# Solution 1: create a scatter plot with Matplotlib
# The x-axis represent the product price, 
# the y-axis would represent the number of views.

# create a scatter plot of product price vs. views
plt.scatter(etsydata['product_price'], etsydata['views'])

# add axis labels and a title
plt.xlabel('Product Price')
plt.ylabel('Number of Views')
plt.title('Product Price vs. Views')

# display the plot
plt.show()

# Conclusion: lower is the price, higher is the number of views.
```

```python
    # Demonstrate how the price affects the number of views
# Solution 2: create a boxplot with Seaborn
# Boxplot show the distribution of views for different price ranges. 
# Show the median, quartiles, and outliers of the views data for each price range. 
# Give an idea of how the number of views varies across different price ranges.

# create a boxplot of views grouped by price range
price_ranges = pd.cut(etsydata['product_price'], bins=[0, 10, 20, 30, 40, 50, 100])
sns.boxplot(x=price_ranges, y=etsydata['views'])

# add axis labels and a title
plt.xlabel('Price Range')
plt.ylabel('Number of Views')
plt.title('Views by Price Range')

# display the plot
plt.show()
```

```python
    # Demonstrate how the price affects the number of views
# Solution 3: create an histogram with matplotlib
# Histogram shows the distribution of product prices and views separately. 
# Give an idea of the most common price and view ranges in the data.

# create a histogram of product prices
# the bins parameter specifies the number of intervals (or "bins") that the range of data values will be divided into
plt.hist(etsydata['product_price'], bins=50)

# add axis labels and a title
plt.xlabel('Product Price')
plt.ylabel('Frequency')
plt.title('Product Price Distribution')

# display the plot
plt.show()

# create a histogram of views
plt.hist(etsydata['views'], bins=50)

# add axis labels and a title
plt.xlabel('Number of Views')
plt.ylabel('Frequency')
plt.title('Views Distribution')

# display the plot
plt.show()
```

```python
    # Demonstrate how the price affects the number of views
# Solution 4: create a heatmap with matplotlib
# Heatmap shows the correlation between product price and views. 
# When the correlation coefficient is close to 0, it means that there is little or no linear relationship between the two variables. 
# In other words, the values of one variable do not predict or explain the values of the other variable, and vice versa.
# The absence of a correlation does not necessarily imply the absence of a relationship between the variables. 
# It could be that the relationship between the variables is non-linear, or that there are other factors that affect the relationship.

# create a heatmap of the correlation between price and views
corr = etsydata[['product_price', 'views', 'avg_review', 'favorites']].corr()
sns.heatmap(corr, annot=True)

# add a title
plt.title('Correlation between Price and Views')

# display the plot
plt.show()
```
